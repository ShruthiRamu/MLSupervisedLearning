```python
# initial imports
import pandas as pd
from path import Path
```

```python
# Load data
file_path = Path("../Resources/loans_data_encoded.csv")
encoded_df = pd.read_csv(file_path)
encoded_df.head()
```

## Scaling Data

```python
# Creating the scaler instance
from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler()
```

```python
# Fitting the scaler
loans_data_scaled = data_scaler.fit_transform(encoded_df)
loans_data_scaled[:5]
```

```python
import numpy as np
print(np.mean(loans_data_scaled[:,0]))
print(np.std(loans_data_scaled[:,0]))
```

```python
rows = 500
columns = 11
for j in range(columns):
     for i in range(rows):
        print(np.mean(loans_data_scaled[:,j]))
        print(np.std(loans_data_scaled[:,j]))
```
